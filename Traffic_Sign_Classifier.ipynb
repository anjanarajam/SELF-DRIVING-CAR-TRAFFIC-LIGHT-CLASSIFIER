### Model Architecture

````python

hidden_layer_depth_1 = 32
# hidden_layer_depth_2 = int(hidden_nodes_1 * 0.5) 
# hidden_layer_depth_3 = int(hidden_nodes_1 * np.power(0.5, 2))
# hidden_layer_depth_4 = int(hidden_nodes_1 * np.power(0.5, 3))
# hidden_layer_depth_5 = int(hidden_nodes_1 * np.power(0.5, 4))

def LeNet(x):
    # Arguments used for tf.truncated_normal, randomly defines variables for the weights and biases for each layer
    # with fine tuning
    mu = 0
    sigma = 0.1
    weights = []
    
    # Layer 1: Convolutional. 
    ## w has a 5x5 filter with an input depth of 3 and output depth of 6       
    weight_1 = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, hidden_layer_depth_1), mean = mu, stddev = sigma), name='weights_0')
    bias_1 = tf.Variable(tf.zeros(hidden_layer_depth_1), name='bias_0')
    weights.append(weight_1)
    
    ## strides = [stride of batch, stride of height, stride of width, stride of feature]
    ## stride of height and width is by default 1
    # Input = 32x32x3. Output = 28x28x32
    logit_1   = tf.nn.conv2d(x, weight_1, strides=[1, 1, 1, 1], padding='VALID') + bias_1
    print("layer1 shape", logit_1.shape)
        
    # Activation.
    hidden_layer_1 = tf.nn.relu(logit_1)
    
    # Pooling. Input = 28x28x32, Output = 14x14x32
    maxpool_hidden_layer_1 = tf.nn.max_pool(hidden_layer_1, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    print("layer 1 shape after max pool", maxpool_hidden_layer_1.shape)
    
    # dropout. Input = 14x14x32, Output = 14x14x32
    dropout_hidden_layer_1 = tf.nn.dropout(maxpool_hidden_layer_1, keep_prob)
    print("layer 1 shape after dropout:", dropout_hidden_layer_1.shape)
    
    # Layer 2: Convolutional. Input = 14x14x32, Output = 10x10x64
    weight_2 = tf.Variable(tf.truncated_normal(shape=(5, 5, 32, 64), mean = mu, stddev = sigma), name='weights_1')
    bias_2 = tf.Variable(tf.zeros(64), name='bias_1')
    logit_2   = tf.nn.conv2d(dropout_hidden_layer_1, weight_2, strides=[1, 1, 1, 1], padding='VALID') + bias_2
    weights.append(weight_2)
    print("layer 2 shape", logit_2.shape)
    
    # Activation.
    hidden_layer_2 = tf.nn.relu(logit_2)

    # Pooling. Input = 10x10x64. Output = 5x5x64
    maxpool_hidden_layer_2 = tf.nn.max_pool(hidden_layer_2, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    print("layer 2 shape after max pool", maxpool_hidden_layer_2.shape)
    
    # dropout. Input = 5x5x64, Output = 5x5x64
    dropout_hidden_layer_2 = tf.nn.dropout(maxpool_hidden_layer_2, keep_prob)
    print("layer 2 shape after dropout:", dropout_hidden_layer_2.shape)
    
#     # Layer 3: Convolutional. Input = 14x14x16, Output = 10x10x16
#     weight_3 = tf.Variable(tf.truncated_normal(shape=(5, 5, 64, 128), mean = mu, stddev = sigma), name='weights_2')
#     bias_3 = tf.Variable(tf.zeros(128), name='bias_2')
#     logit_3   = tf.nn.conv2d(dropout_hidden_layer_2, weight_3, strides=[1, 1, 1, 1], padding='VALID') + bias_3
#     weights.append(weight_3)
#     print("layer 3 shape", logit_3.shape)
    
#     # Activation.
#     hidden_layer_3 = tf.nn.relu(logit_3)

    # Flatten. Input = 5x5x16. Output = 1024.
    fc0 = flatten(dropout_hidden_layer_2)
    print("fc0 shape", fc0.shape)
    
    # Layer 4: Fully Connected. Input = 1024. Output = 122.
    weight_4 = tf.Variable(tf.truncated_normal(shape=(1600, 512), mean = mu, stddev = sigma), name='weights_3')
    bias_4 = tf.Variable(tf.zeros(512), name='bias_3')
    logit_4   = tf.matmul(fc0, weight_4) + bias_4
    weights.append(weight_4)
    print("layer 4 shape:", logit_4.shape)
    
    # Activation.
    hidden_layer_4  = tf.nn.relu(logit_4)
    
    # dropout. Input = 28x28x6, Output = 14x14x16
    dropout_hidden_layer_4 = tf.nn.dropout(hidden_layer_4, keep_prob)
    print("layer 4 shape after dropout:", dropout_hidden_layer_4.shape)

    # Layer 5: Fully Connected. Input = 120. Output = 84.
    weight_5  = tf.Variable(tf.truncated_normal(shape=(512, 256), mean = mu, stddev = sigma), name='weights_4')
    bias_5  = tf.Variable(tf.zeros(256), name='bias_4')
    logit_5  = tf.matmul(dropout_hidden_layer_4, weight_5) + bias_5
    weights.append(weight_5)
    
    # Activation.
    hidden_layer_5 = tf.nn.relu(logit_5)
    
    # dropout. Input = 28x28x6, Output = 14x14x16
    dropout_hidden_layer_5 = tf.nn.dropout(hidden_layer_5, keep_prob)
    print("layer 5 shape after dropout:", dropout_hidden_layer_5.shape)

    # Layer 6: Fully Connected. Input = 84. Output = 10.
    weight_6  = tf.Variable(tf.truncated_normal(shape=(256, 125), mean = mu, stddev = sigma), name='weights_5')
    bias_6  = tf.Variable(tf.zeros(125), name='bias_5')
    logit_6 = tf.matmul(dropout_hidden_layer_5, weight_6) + bias_6
    weights.append(weight_6)
    
    # Activation.
    hidden_layer_6 = tf.nn.relu(logit_6)
    
    # dropout. Input = 28x28x6, Output = 14x14x16
    dropout_hidden_layer_6 = tf.nn.dropout(hidden_layer_6, keep_prob)
    print("layer 6 shape after dropout:", dropout_hidden_layer_6.shape)
    
    # Layer 7: Fully Connected. Input = 84. Output = 10.
    weight_7  = tf.Variable(tf.truncated_normal(shape=(125, n_classes), mean = mu, stddev = sigma), name='weights_6')
    bias_7  = tf.Variable(tf.zeros(n_classes), name='bias_6')
    logit_7 = tf.matmul(dropout_hidden_layer_6, weight_7) + bias_7
    weights.append(weight_7)
    
        
    return logit_7, weights
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
# Number od channels is 1 since it is a grayscale image
x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32)
one_hot_y = tf.one_hot(y, n_classes)
# 0.001 is a good default value
learning_rate = 0.001
# Initialize regularizer
regularizer = 0
# beta for regularizer
beta = 0.001

# Get the logits from the architecture
logits, weights = LeNet(x)
# cross entropy means how different the logits are from groud truth labels
cross_entropy = tf.nn.softmax_cross_entropy_with_logits(labels=one_hot_y, logits=logits)
# loss function is obtained by averaging cross entropy
loss_operation = tf.reduce_mean(cross_entropy)
# Loss function using L2 Regularization
for weight in range(len(weights)):
    regularizer += tf.nn.l2_loss(weights[weight])
loss_operation = tf.reduce_mean(loss_operation + beta * regularizer)

# Deccaying learning rate for the optimizer
# global_step to count the number of steps taken.
global_step = tf.Variable(0)  
learning_rate = tf.train.exponential_decay(learning_rate, global_step, 100000, 0.96, staircase=True)
# To minimise the loss function, adams algorithm is uses similar 
# to gradient descent algorithm.we run the minimise function on the optimiser which uses
# backpropagation to update the network and minimise the training loss
optimizer = tf.train.AdamOptimizer(learning_rate)
training_operation = optimizer.minimize(loss_operation, global_step=global_step)

### Calculate and report the accuracy on the btraining and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.
````

**Accuracy of traning set**

````python
EPOCHS = 20
BATCH_SIZE = 256

## Measure whether if a given prediction is correct by comparing the logit prediction to the one hot
## encoded groundtruth label
correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
## Second step is to calculate the models overall accuracy by averaging the individual
## predicion accuracies
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

## in order to run the evaluation pipelione, we have to build an evaluate function. This function takes data set as inputs,
## sets some initial values, batches it and then runs them 
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.5})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples

# Run the training data through the training pipeline to train the model.
with tf.Session() as sess:
    sess.run(tf.global_variables_initializer())
    num_examples = len(X_train)
    
    print("Training...")
    print()
    for i in range(EPOCHS):
        X_train, y_train = shuffle(X_train, y_train)
        for offset in range(0, num_examples, BATCH_SIZE):
            end = offset + BATCH_SIZE
            batch_x, batch_y = X_train[offset:end], y_train[offset:end]
            sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
            
        validation_accuracy = evaluate(X_valid, y_valid)
        print("EPOCH {} ...".format(i+1))
        print("Validation Accuracy = {:.3f}".format(validation_accuracy))
        print()
        
    saver.save(sess, './lenet')
    print("Model saved")
````
